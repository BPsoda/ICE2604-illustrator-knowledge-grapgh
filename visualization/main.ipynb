```python
from bs4 import BeautifulSoup
with open("466数字标签.svg","r",encoding="utf-8") as f:
    a=f.read()
a=BeautifulSoup(a,"xml")
for i in a.find_all("circle"):
    print(i)
    break
```

```python
pos=[[i["class"],float(i["cx"]),float(i["cy"])] for i in a.find_all("circle")]
```

```python
import matplotlib.pyplot as plt
import numpy as np
b=np.array(pos)
print(b[0])
plt.scatter(b[:,1],b[:,2])
plt.show()
```

```python
from 聚类 import  K_means
import random
c=K_means(pos,[i[1:] for i in random.sample(pos,10)])
```

```python
import matplotlib.pyplot as plt
for i in c:
    plt.scatter([j[1] for j in i],[j[2] for j in i])
plt.show()
```

```python
for i in c:
    for j in i:
        print(j[0],end=" ")
    print()
```
